# Explication
* L'ensemble de données contient des transactions effectuées par des cartes de crédit en septembre 2013 par des titulaires de cartes européens.

* Cet ensemble de données présente des transactions qui ont eu lieu sur deux jours, avec 492 fraudes sur un total de 284 807 transactions.

* L'ensemble de données est très déséquilibré, la classe positive (fraudes) représente 0,172 % de l'ensemble des transactions.

* Il ne contient que des variables d'entrée numériques qui sont le résultat d'une transformation en composantes principales (PCA).

* Malheureusement, en raison de problèmes de confidentialité, nous ne pouvons pas fournir les caractéristiques originales et davantage d'informations sur les antécédents des données.

* Les caractéristiques V1, V2, … V28 sont les composantes principales obtenues avec ACP, les seules caractéristiques qui n'ont pas été transformées avec PCA sont 'Time' (Temps) et 'Amount' (Montant).

* La caractéristique 'Time' contient les secondes écoulées entre chaque transaction et la première transaction dans l'ensemble de données.

* La caractéristique 'Amount' est le montant de la transaction. Cette caractéristique peut être utilisée, par exemple, pour l'apprentissage dépendant du coût.

* La caractéristique 'Class' est la variable de réponse et elle prend la valeur 1 en cas de fraude et 0 sinon. L'ensemble de données contient un pourcentage très faible de transactions frauduleuses. Nous devons identifier ces transactions qui appartiennent à la classe Fraud (Fraude).

* Sur la base des données, nous devons générer un ensemble d'observations et de recommandations qui aideront la société de cartes de crédit à prévenir les débits frauduleux pour ses clients.

```python
columns = data.columns.tolist()
columns = [c for c in columns if c not in ['Class']]

target = 'Class'

X = data[columns]
Y = data[target]


print(X.shape)
print(Y.shape)
```

```python
from sklearn.metrics import classification_report, accuracy_score
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
```

```python
# determine the number of fraud cases
fraud = data[data['Class'] == 1]
valid = data[data['Class'] == 0]

outlier_fraction = len(fraud) / float(len(valid))
print(outlier_fraction)

print('Fraud Cases: {}'.format(len(fraud)))
print('Valid Cases: {}'.format(len(valid)))
```

```python
state = 1

# definir l'outlier detection methode
classifiers = {
    # contamination est le nombre d' outliers qu'il y'a
    'Isolation Forest': IsolationForest(max_samples = len(X),
                                       contamination = outlier_fraction,
                                       random_state = state),
    # nombfre de voisins
    'Local Outlier Factor': LocalOutlierFactor(
    n_neighbors = 20,
    contamination = outlier_fraction)
}
```

```python
n_outliers = len(fraud)

for i, (clf_name, clf) in enumerate(classifiers.items()):

    # fit the data
    if clf_name == 'Local Outlier Factor':
        y_pred = clf.fit_predict(X)
        scores_pred = clf.negative_outlier_factor_
    else:
        clf.fit(X)
        scores_pred = clf.decision_function(X)
        y_pred = clf.predict(X)



   # 0 pour valid et 1 pour fraud
    y_pred[y_pred == 1] = 0
    y_pred[y_pred == -1] = 1

    #  nombre d"erreurs
    n_errors = (y_pred != Y).sum()

    # matrice de classification
    print('{}: {}'.format(clf_name, n_errors))
    print(accuracy_score(Y, y_pred))
    print(classification_report(Y, y_pred))
```
