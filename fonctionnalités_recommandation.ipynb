```python
import pandas as pd
from faker import Faker
import random
from datetime import datetime, timedelta

fake = Faker('fr_FR')

donnees = []

utilisateurs = []
for _ in range(100):
    nom = fake.first_name()
    prenom = fake.last_name()
    utilisateurs.append(f"{nom} {prenom}")

types_de_transactions = ['Achat en ligne', 'Retrait au distributeur', 'Paiement en magasin', 'Virement bancaire']
heures_de_transaction = ['08:00', '12:30', '15:45', '18:20', '21:10']

for _ in range(1000):
    utilisateur = random.choice(utilisateurs)
    adresse = fake.street_address()
    ville = fake.city()
    code_postal = fake.postcode()
    email = fake.email()
    numero_telephone = fake.phone_number()
    numero_carte_credit = fake.credit_card_number(card_type='mastercard')
    date_expiration_carte = fake.credit_card_expire()
    salaire_annuel = fake.random_int(min=20000, max=200000)
    montant_transaction = fake.random_int(min=10, max=50000)

    fraude = random.choice([0, 0, 0, 1])

    investissement_1 = random.choice(['Actions', 'Obligations', 'Immobilier', 'Fonds indiciels'])
    investissement_2 = random.choice(['Actions', 'Obligations', 'Immobilier', 'Fonds indiciels'])
    investissement_3 = random.choice(['Actions', 'Obligations', 'Immobilier', 'Fonds indiciels'])

    heure_transaction = random.choice(heures_de_transaction)
    type_transaction = random.choice(types_de_transactions)

    # Génération d'une date de transaction aléatoire dans les 10 premiers jours du mois en cours
    date_transaction = fake.date_between_dates(date_start=datetime.today().replace(day=1), date_end=datetime.today().replace(day=10))
    date_fin_mois = date_transaction + timedelta(days=20)
    date_transaction_str = date_transaction.strftime('%Y-%m-%d')

    flux_tresorerie = salaire_annuel - montant_transaction

    donnees.append([utilisateur, adresse, ville, code_postal, email, numero_telephone,
                    numero_carte_credit, date_expiration_carte, salaire_annuel, montant_transaction, date_transaction_str,
                    investissement_1, investissement_2, investissement_3, heure_transaction, type_transaction,
                    flux_tresorerie, fraude])

df = pd.DataFrame(donnees, columns=["Utilisateur", "Adresse", "Ville", "Code Postal", "Email",
                                     "Numéro de Téléphone", "Numéro de Carte de Crédit", "Date d'Expiration de la Carte",
                                     "Salaire Annuel", "Montant de la Transaction", "Date de la Transaction", "Investissement 1",
                                     "Investissement 2", "Investissement 3", "Heure de la Transaction", "Type de Transaction",
                                     "Flux de Trésorerie", "Fraude"])

df.to_csv('dataset_algerie.csv', index=False)
```

```python
#Importe la fonction cosine_similarity de la bibliothèque sklearn.metrics.pairwise, qui sera utilisée pour calculer la similarité cosinus entre les utilisateurs.
from sklearn.metrics.pairwise import cosine_similarity

data = pd.read_csv('dataset_algerie.csv')


data = data.drop_duplicates(subset='Utilisateur')

#Crée une matrice où les lignes représentent les utilisateurs, les colonnes représentent les types d'investissements ('Investissement 1' dans ce cas), et les valeurs sont les montants de transaction correspondants. Les valeurs manquantes sont remplies avec des zéros.
utility_matrix = data.pivot(index='Utilisateur', columns='Investissement 1', values='Montant de la Transaction').fillna(0)


user_similarity = cosine_similarity(utility_matrix)


recommended_investments = {}
for user in utility_matrix.index:
    user_index = utility_matrix.index.get_loc(user)
    similar_users_indices = user_similarity[user_index].argsort()[:-1]


    recommended_investments[user] = []
    for similar_user_index in similar_users_indices:
        similar_user_row = utility_matrix.iloc[similar_user_index]
        recommended_investments[user].extend(similar_user_row[similar_user_row > 0].index)


    current_investments = utility_matrix.loc[user]
    recommended_investments[user] = [inv for inv in recommended_investments[user] if pd.isna(current_investments[inv])]


for user, investments in recommended_investments.items():
    print(f"Investissements recommandés pour {user} :", investments)
```
